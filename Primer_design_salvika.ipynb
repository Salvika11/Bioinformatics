```python
primer_length_min = 18
primer_length_max = 24
gc_content_min = 40
gc_content_max = 60
tm_min = 55
tm_max = 65
product_size_min = 100
product_size_max = 1000

print(f"Primer length range: {primer_length_min}-{primer_length_max} bp")
print(f"GC content range: {gc_content_min}-{gc_content_max}%")
print(f"Melting temperature range: {tm_min}-{tm_max}°C")
print(f"Product size range: {product_size_min}-{product_size_max} bp")

def calculate_gc_content(sequence):
    """
    Calculates the GC content of a DNA sequence.
    Args:
        sequence (str): The DNA sequence.
    Returns:
        float: The GC content as a percentage.
    """
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    gc_content = ((g_count + c_count) / len(sequence)) * 100
    return gc_content

def calculate_tm(sequence):
    """
    Calculates the simplified melting temperature (Tm) of a DNA sequence.
    Args:
        sequence (str): The DNA sequence.
    Returns:
        float: The melting temperature in Celsius.
    """
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    a_count = sequence.count('A')
    t_count = sequence.count('T')
    tm = (4 * (g_count + c_count)) + (2 * (a_count + t_count))
    return tm

print("Functions `calculate_gc_content` and `calculate_tm` have been defined.")
```

```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import io

# --- IMPORTANT --- #
# This code assumes `fasta_sequence_data` contains a valid FASTA sequence.
# If it still contains the placeholder 'YOUR_FASTA_SEQUENCE_HERE',
# this step will execute but subsequent primer calculations will fail.

# Extract the primary sequence
# Use io.StringIO to parse the string as a file-like object
# Using 'fasta-pearson' format to handle potential comments/whitespace more robustly, as per BiopythonDeprecationWarning
sequence_file = io.StringIO(fasta_sequence_data)
try:
    record = SeqIO.read(sequence_file, "fasta-pearson")
    primary_sequence = str(record.seq).upper() # Ensure uppercase for consistency
except ValueError as e:
    print(f"Error parsing FASTA sequence: {e}")
    print("Please ensure 'fasta_sequence_data' contains a valid FASTA format, including the header and sequence.")
    primary_sequence = ""

# Generate the reverse complement sequence
reverse_complement_sequence = ""
if primary_sequence:
    reverse_complement_sequence = str(record.seq.reverse_complement()).upper()

print(f"Primary sequence extracted (first 50 chars): {primary_sequence[:50]}...")
print(f"Reverse complement sequence generated (first 50 chars): {reverse_complement_sequence[:50]}...")

# 3. Initialize empty lists for forward_primer_candidates and reverse_primer_candidates.
forward_primer_candidates = []
reverse_primer_candidates = []

# Helper function for primer filtering
def is_valid_primer(sequence, gc_min, gc_max, tm_min, tm_max):
    if not sequence:
        return False
    gc = calculate_gc_content(sequence)
    tm = calculate_tm(sequence)
    return (gc_min <= gc <= gc_max) and (tm_min <= tm <= tm_max)

# 4. Generate and filter forward primer candidates
if primary_sequence:
    for i in range(len(primary_sequence)):
        for length in range(primer_length_min, primer_length_max + 1):
            if i + length <= len(primary_sequence):
                candidate = primary_sequence[i : i + length]
                if is_valid_primer(candidate, gc_content_min, gc_content_max, tm_min, tm_max):
                    forward_primer_candidates.append({
                        'sequence': candidate,
                        'length': length,
                        'gc': calculate_gc_content(candidate),
                        'tm': calculate_tm(candidate),
                        'start_index': i
                    })
print(f"Found {len(forward_primer_candidates)} forward primer candidates.")

# 5. Generate and filter reverse primer candidates
if reverse_complement_sequence:
    for i in range(len(reverse_complement_sequence)):
        for length in range(primer_length_min, primer_length_max + 1):
            if i + length <= len(reverse_complement_sequence):
                candidate = reverse_complement_sequence[i : i + length]
                if is_valid_primer(candidate, gc_content_min, gc_content_max, tm_min, tm_max):
                    # Calculate the 3' end index on the primary sequence for the reverse primer
                    # If RP is at reverse_complement_sequence[i:i+length], its 3' end binds to primary_sequence[len(primary_sequence) - 1 - i]
                    rp_end_index_on_primary = len(primary_sequence) - 1 - i
                    reverse_primer_candidates.append({
                        'sequence': candidate,
                        'length': length,
                        'gc': calculate_gc_content(candidate),
                        'tm': calculate_tm(candidate),
                        'end_index_on_primary': rp_end_index_on_primary
                    })
print(f"Found {len(reverse_primer_candidates)} reverse primer candidates.")

# 6. Initialize an empty list called potential_primer_pairs.
potential_primer_pairs = []

# 7-10. Iterate through each forward primer and reverse primer to form pairs
#       and filter by product size.
for fp in forward_primer_candidates:
    for rp in reverse_primer_candidates:
        fp_start_index = fp['start_index']
        rp_end_index_on_primary = rp['end_index_on_primary']

        # Ensure the reverse primer binds after the forward primer
        if rp_end_index_on_primary > fp_start_index:
            product_size = rp_end_index_on_primary - fp_start_index + 1

            # Filter primer pairs based on the product_size_min and product_size_max range.
            if product_size_min <= product_size <= product_size_max:
                potential_primer_pairs.append({
                    'Forward_Primer': fp['sequence'],
                    'FP_Length': fp['length'],
                    'FP_GC': f"{fp['gc']:.2f}%",
                    'FP_Tm': f"{fp['tm']:.2f} ℃",
                    'FP_Start_Index': fp_start_index,
                    'Reverse_Primer': rp['sequence'],
                    'RP_Length': rp['length'],
                    'RP_GC': f"{rp['gc']:.2f}%",
                    'RP_Tm': f"{rp['tm']:.2f} ℃",
                    'RP_End_Index': rp_end_index_on_primary,
                    'Product_Size': product_size
                })

print(f"Found {len(potential_primer_pairs)} potential primer pairs.")
import pandas as pd

# Convert the list of potential primer pairs into a DataFrame for better presentation
if potential_primer_pairs:
    df_primer_results = pd.DataFrame(potential_primer_pairs)
    print(f"Successfully generated a DataFrame with {len(df_primer_results)} potential primer pairs.")
    print("Preview of potential primer pairs:")
    print(df_primer_results.head())
else:
    df_primer_results = pd.DataFrame(columns=[
        'Forward_Primer', 'FP_Length', 'FP_GC', 'FP_Tm', 'FP_Start_Index',
        'Reverse_Primer', 'RP_Length', 'RP_GC', 'RP_Tm', 'RP_End_Index',
        'Product_Size'
    ])
    print("No potential primer pairs found. Please ensure the FASTA sequence is valid and parameters are appropriate.")

# The subtask of generating, filtering, and pairing primers is now complete.
# The results are stored in `df_primer_results` (which might be empty).
```
